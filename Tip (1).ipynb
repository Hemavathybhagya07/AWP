```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("tips.csv")
display(data.head(10))
```

```python
# reading the database
data = pd.read_csv("tips.csv")
 
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'], c=data['size'], 
            s=data['total_bill'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.colorbar()
 
plt.show()
```

```python
#creating the barplot
#x='day': The x-axis represents the days.
#y='tip': The y-axis represents the tips.
#data=data: The dataset being used is named data.
#hue='sex': The bars are colored based on the sex column in the dataset.

sns.barplot(x='day',y='tip', data=data, 
            hue='sex')
 
plt.show()
```

```python
#total bill according to the gender

sns.histplot(x='total_bill', data=data, kde=True, hue='sex')
 
plt.show()                       #KDE tocreate a density curve
```

```python
sns.barplot(x='sex',y='smoker',data=data)
plt.xlabel('sex')
plt.ylabel('smoker')
plt.savefig('tips')
plt.show()
```
